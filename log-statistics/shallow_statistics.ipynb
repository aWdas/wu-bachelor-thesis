# Wikidata

# All

```python
import seaborn as sns
import pandas as pd

stats = pd.DataFrame(columns=["interval", "lines", "total", "valid", "variable_pred", "subquery", "pred_path"])

intervals = list(range(1, 8))

for i in intervals:
    lines = open(f"../queryshapes_wikidata_all_{i}.tsv").readlines()
    i_stats = [int(line.split("\t")[1]) for line in lines[:6]]
    stats = stats.append({"interval": i, "lines": i_stats[0], "total": i_stats[1], "valid": i_stats[2], "variable_pred": i_stats[3],
                  "subquery": i_stats[4], "pred_path": i_stats[5]}, ignore_index=True)
    
stats["variable_pred_pct"] = stats["variable_pred"] / stats["valid"] * 100
stats["subquery_pct"] = stats["subquery"] / stats["valid"] * 100
stats["pred_path_pct"] = stats["pred_path"] / stats["valid"] * 100
stats["valid_pct"] = stats["valid"] / stats["total"] * 100
stats = stats.append({
    "interval": 0,
    "lines": stats["lines"].sum(),
    "total": stats["total"].sum(),
    "valid": stats["valid"].sum(),
    "valid_pct": stats["valid"].sum() / stats["total"].sum() * 100,
    "variable_pred_pct": stats["variable_pred"].sum() / stats["valid"].sum() * 100,
    "subquery_pct": stats["subquery"].sum() / stats["valid"].sum() * 100,
    "pred_path_pct": stats["pred_path"].sum() / stats["valid"].sum() * 100,
}, ignore_index=True)
```

```python
stats
```
